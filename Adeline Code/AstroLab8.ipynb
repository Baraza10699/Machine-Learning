```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([40,80,125,210,298,340,420,560,680,869])
y =np.array([20,40,65,100,153,172,200,250,300,350])
x_c = np.array([420,560,680,860])
y_c = np.array([200,250,300,350])
```

```python
m,b = np.polyfit(x[:7],y[:7],1)
```

```python
fig,ax = plt.subplots(figsize =(8,10))

ax.set_xlim (0,900)
ax.set_title('Velocity vs Distance')

ax.title.set_fontsize(20)
ax.set_xlabel('Distance(mil)')
ax.xaxis.label.set_fontsize(20)
ax.set_ylabel('Velocity(mph)')
ax.yaxis.label.set_fontsize(20)

ax.scatter(x,y)
ax.scatter(x_c,y_c,color ='orange')
ax.set_xticks(range(0,900,100))
ax.tick_params(axis ='both',labelsize = 14)
plt.plot(x,m *x +b,color ='purple',label='Cars 1-6')
plt.legend(loc ='upper center')
```
